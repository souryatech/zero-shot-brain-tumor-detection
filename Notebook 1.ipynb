## Image Processing

Find size of images; make sure images are all same size; convert images to RGB (I checked it was already RGB and not grayscale).

```
import os
from PIL import Image

image_files = sorted([f for f in os.listdir(image_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
print(f"Found {len(image_files)} images.")

#first img shape
first_img = Image.open(os.path.join(image_folder, image_files[0])).convert('RGB')
# img_width, img_height = first_img.size
img_width = 150
img_height = 150
print("width: " + str(img_width) + " height: " + str(img_height))

```

## 

```
num_images = len(image_files)
X_train = np.zeros((num_images, img_height, img_width, 3), dtype=np.uint8)

for i, img_name in enumerate(image_files):
    img_path = os.path.join(image_folder, img_name)
    img = Image.open(img_path).convert('RGB')
    img = img.resize((img_width, img_height))  
    X_train[i] = np.array(img)

print(f"Loaded images into NumPy array with shape {X_train.shape}")
```

```
#Normalize dat shi 
X_train = X_train.astype(np.float32) / 255.0
```
